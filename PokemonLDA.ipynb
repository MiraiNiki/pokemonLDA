```python
import numpy as np
import pandas as pd

df = pd.read_csv("Data/name_type.csv")
df
```

```python
poke = {}
cnt = 0
for row in df.itertuples():
    if row[1] not in poke:
        poke[row[1]] = cnt
        cnt = cnt + 1
poke
```

```python
tlist = [[] for i in range(len(poke))]
for row in df.itertuples():
    tlist[poke[row[1]]].append(row[2])
tlist
```

```python
top = [[] for i in range(5)]
hot = [[] for i in range(5)]
for topic in lda5[type_corpus]:
    for tup in topic:
        top[tup[0]].append(tup[1])
        hot[tup[0]].append(tup[1])
        
for i in range(5):
    index = len(poke) - len(top[i])
    for j in range(index):
        top[i].append(0)
len(top[2])
```
