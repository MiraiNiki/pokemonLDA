```python
from gensim.test.utils import common_texts
from gensim.corpora.dictionary import Dictionary

common_dictionary = Dictionary(common_texts)
common_corpus = [common_dictionary.doc2bow(text) for text in common_texts]
common_texts
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv("Data/name_type.csv")
df
```

```python
poke = {}
cnt = 0
for row in df.itertuples():
    if row[1] not in poke:
        poke[row[1]] = cnt
        cnt = cnt + 1
poke
```

```python
tlist = [[] for i in range(len(poke))]
for row in df.itertuples():
    tlist[poke[row[1]]].append(row[2])
tlist
```

```python
from gensim.test.utils import common_texts
from gensim.corpora.dictionary import Dictionary
type_dictionary = Dictionary(tlist)
type_corpus = [type_dictionary.doc2bow(t) for t in tlist]
type_corpus
```

```python
import gensim.models.ldamodel as gs
lda = gensim.models.ldamodel.LdaModel(corpus=type_corpus, num_topics=10, id2word=type_dictionary)
lda.show_topics()
```
